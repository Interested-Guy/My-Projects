```python
import pandas as pd             #Get the data as a dataframe
df=pd.read_csv("score.csv")
df
```

```python
df.corr()                       #Observe relationship between two variables
```

```python
import statsmodels.api as sm    #Preparing model
x=df['Hours']
y=df['Scores']
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
model.summary()
```

##The model is basically
###y=2.4837+9.7758x

```python
l=model.params        #Model parameters
print(l[0])           #Y intercept
print(l[1])           #The slope of the regression line
```

```python
#Function to predict based on score
def predict(n):
  y=l[0]+l[1]*n
  return y
print("Score when no. of hours=9.25 is:",int(predict(9.25)))
```
