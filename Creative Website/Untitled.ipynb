```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("project.csv");
def f(s):
    for i in range(len(s)):
        if(s[i]=='1'):
            return 1
df['fruit']=df.crop.apply(f)
def f1(s):
    for i in range(len(s)):
        if(s[i]=='2'):
            return 1
df['vegetable']=df.crop.apply(f1)        
def f2(s):
    for i in range(len(s)):
        if(s[i]=='3'):
            return 1
df['legume']=df.crop.apply(f2)
def f3(s):
    for i in range(len(s)):
        if(s[i]=='4'):
            return 1
df['grain']=df.crop.apply(f3)
l1=[]
l1.append(int(df.fruit.value_counts()))
l1.append(int(df.vegetable.value_counts()))
l1.append(int(df.legume.value_counts()))
l1.append(int(df.grain.value_counts()))
plt.bar(["fruit","vegetable","legume","grain"],l1)
plt.show()
```
